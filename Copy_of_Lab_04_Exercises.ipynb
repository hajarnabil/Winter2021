## Exercise 1

Let's load lab 3's dataset again and try to handle the missing values this time.

```python
try:
    world_df = pd.read_csv('../data/world_factbook.csv')
except FileNotFoundError:
    world_df = pd.read_csv('https://raw.githubusercontent.com/GUC-DM/W2021/main/data/world_factbook.csv')

world_df.head()
```

```python
# instead of the automaticaly generated index, let's use the country column
world_df = world_df.set_index('country')
world_df.head()
```

**How many missing values are there in each column?**

```python
world_df.isnull().sum()
```

**Filter the data to the rows where there are missing values**

_Hint: save the boolean mask to a variable for re-use in later prompts._

```python
world_df.isnull().head()
```

```python
world_df.isnull().sum(axis=1)
```

```python
missing = world_df.isnull().sum(axis=1) > 0
missing
```

```python
world_df[missing]
```

**Fill the missing values with the dataset's mean for each of the columns and assign the result to a new variable `world_df_na_mean`. How does the data filtered above look now?**

```python
world_df.mean()
```

```python
world_df_na_mean = world_df.fillna(world_df.mean())
world_df_na_mean[missing].head()
```

```python
world_df.columns
```

```python
cols = ['migration', 'infant_mortality', 'gdp', 'literacy', 'phones', 'arable', 'crops',
                'other', 'birthrate', 'deathrate', 'agriculture', 'industry', 'service']

world_df_na_mean = world_df.fillna(world_df[cols].mean())
world_df_na_mean[missing].head()
```

```python
world_df_na_mean['climate'].fillna(world_df_na_mean['climate'].mode()[0], inplace=True)
world_df_na_mean.head()
```

**Fill the missing values based on each region's median for each of the columns and assign the result to a new variable `world_df_na_median`. How does the data filtered earlier look now? What is the advantage of this approach?**

_Hint: you'll need to use `groupby`_

```python
world_df.groupby('region').median()
```

```python
world_df_na_median = world_df.groupby('region').apply(lambda x: x.fillna(x.median()))
world_df_na_median
```

```python
world_df_na_median.droplevel('region')[missing].sort_index().head()
```

```python
cols = ['migration', 'infant_mortality', 'gdp', 'literacy', 'phones', 'arable', 'crops',
        'other', 'birthrate', 'deathrate', 'agriculture', 'industry', 'service']
world_df_na_median = world_df.copy()
world_df_na_median[cols] = world_df_na_median.groupby('region')[cols].apply(lambda x: x.fillna(x.median())).droplevel('region')
world_df_na_median.head()
```

```python
world_df_na_median[['region','climate']].groupby('region').agg(pd.Series.mode)
```

```python
world_df_na_median['climate'] = world_df_na_median.groupby('region')['climate'].apply(lambda x: x.fillna(x.mode()[0]))
world_df_na_median
```

## Exercise 2

The gapminder dataset provides data about the population, life expectancy and GDP in different countries of the world from 1952 to 2007. Let's say one of the tasks you wanted to perform later on is to build a regression model to forecast GDP, population, life expectancy of future years, but to do that you needed to first transform the data provided.

```python
try:
    gapminder_df = pd.read_csv("../data/gapminder.csv")
except FileNotFoundError:
    gapminder_df = pd.read_csv('https://raw.githubusercontent.com/GUC-DM/W2021/main/data/gapminder.csv')

gapminder_df
```

**Transform the data provided. Expected columns in the final dataset:** `['Continent','Country','Year','Population Count','GDP Value', 'Life Expectancy']`

_Hint: you'll need to `melt` each variable on its own and then merge the partial dataframes in the end (consider using `pd.concat` or `join`)._

```python
gdp = gapminder_df.loc[:,'gdpPercap_1952':'gdpPercap_2007']
gdp.columns
```

```python
gdp
```

```python
gdp = pd.melt(gdp, var_name='Year', value_name='GDP Value')
gdp.head()
```

```python
gdp['Year'] = gdp['Year'].str.extract('(\d)')
gdp
```

```python
population = pd.melt(gapminder_df.loc[:,'pop_1952':'pop_2007'], var_name='Year', value_name='Population Count')
population.head()
```

```python
population['Year'] = population['Year'].str.extract('(\d)')
population.head()
```

```python
life_exp = pd.melt(gapminder_df.loc[:,'lifeExp_1952':'lifeExp_2007'], var_name='Year', value_name='Life Expectancy')
life_exp.head()
```

```python
life_exp['Year'] = life_exp['Year'].str.extract('(\d+)')
life_exp.head()
```

```python
gapminder_df_final = pd.melt(gapminder_df.loc[:,'contient':'gdpPercap_2007'], id_vars=['continent','country'])
gapminder_df_final
```

```python
gapminder_df_final = gapminder_df_final.drop(['variable','value'],axis=1)
gapminder_df_final.columns = gapminder_df_final.columns.str.capitalize()
gapminder_df_final.head()
```

```python
gapminder_df_final = pd.concat([gapminder_df_final, population, gdp['GDP Value'], life_exp['Life Expectancy']], axis=1)
gapminder_df_final
```
